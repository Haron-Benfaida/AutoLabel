```python
from datasets import load_dataset
from utils.str_utils import Utils
from llm.EngineLLM import LLMEngine
from visualization.drawer import Drawer
from ocr.TesseractOCR import TesseractOCR
from PIL import Image, ImageDraw, ImageFont
import easyocr
```

```python
dataset = load_dataset("ayoub999/dataset_for_orange_factures")
```

```python
example = dataset['train'][3]['image']
reader = easyocr.Reader(['fr'])
```

```python
result = reader.readtext(example)
```

```python
result[0][1]
```

```python
canvas = ImageDraw.Draw(example)
font = ImageFont.load_default()
for data in result:
    box = data[0][0] + data[0][2]
    canvas.rectangle(box, outline='green')
    canvas.text((box[0] + 10, box[1] - 10), text=data[1], fill='green', font=font)

example
```
